```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import openai
```

```python
def load_csv(file_path):
    df = pd.read_csv(file_path)
    print("First few rows of the dataset:")
    print(df.head())
    return df
```

```python
# Function to perform descriptive statistics
def descriptive_statistics(df, column):
    mean = df[column].mean()
    median = df[column].median()
    mode = df[column].mode()[0]
    std_dev = df[column].std()
    return mean, median, mode, std_dev
```

```python
# Function to generate histogram
def plot_histogram(df, column):
    plt.figure(figsize=(10, 5))
    sns.histplot(df[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Function to generate scatter plot
def plot_scatter(df, column_x, column_y):
    plt.figure(figsize=(10, 5))
    sns.scatterplot(x=df[column_x], y=df[column_y])
    plt.title(f'Scatter Plot of {column_x} vs {column_y}')
    plt.xlabel(column_x)
    plt.ylabel(column_y)
    plt.show()
```

```python
# Function to generate line plot
def plot_line(df, column_x, column_y):
    plt.figure(figsize=(10, 5))
    sns.lineplot(x=df[column_x], y=df[column_y])
    plt.title(f'Line Plot of {column_x} vs {column_y}')
    plt.xlabel(column_x)
    plt.ylabel(column_y)
    plt.show()
```

```python

# Function to interpret prompts using LLM
def interpret_prompt(prompt):
    openai.api_key = 'your-api-key'  # Ensure you replace this with your actual API key
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=100
    )
    return response.choices[0].text.strip()
```

```python
file_path = 'covid19.csv'  # Replace with your CSV file path
df = load_csv(file_path)
column = 'Lat' 
```

```python
# Perform descriptive statistics
mean, median, mode, std_dev = descriptive_statistics(df, column)
print(f"Mean: {mean}, Median: {median}, Mode: {mode}, Std Dev: {std_dev}")
```

```python
# Generate plots
plot_histogram(df, column)
plot_scatter(df,'Lat','Long')  
plot_line(df,'Lat','Long')
```
